# Interactive Procedural Terrain Generation
### Binder-Compatible Jupyter Environment

This notebook demonstrates how to generate procedural 3D terrain using **Perlin Noise** and visualize it interactively using **Plotly** and **ipywidgets**.

```python
import numpy as np
import plotly.graph_objects as go
from noise import pnoise2
import ipywidgets as widgets
from IPython.display import display

print("Libraries imported successfully!")
```

## Terrain Generation Logic

```python
def generate_terrain(size=100, scale=10.0, octaves=6, persistence=0.5, lacunarity=2.0, seed=42):
    """Generates a 2D array of heights using Perlin noise."""
    shape = (size, size)
    terrain = np.zeros(shape)
    
    for i in range(size):
        for j in range(size):
            terrain[i][j] = pnoise2(i/scale, 
                                    j/scale, 
                                    octaves=octaves, 
                                    persistence=persistence, 
                                    lacunarity=lacunarity, 
                                    repeatx=1024, 
                                    repeaty=1024, 
                                    base=seed)
    return terrain
```

## Interactive Visualization

```python
def update_plot(size, scale, octaves, persistence, lacunarity, seed, colorscale):
    z = generate_terrain(size, scale, octaves, persistence, lacunarity, seed)
    
    fig = go.Figure(data=[go.Surface(z=z, colorscale=colorscale)])
    
    fig.update_layout(
        title=f'Procedural Terrain (Seed: {seed})',
        autosize=False,
        width=800, height=800,
        margin=dict(l=65, r=50, b=65, t=90),
        scene=dict(
            xaxis_title='X',
            yaxis_title='Y',
            zaxis_title='Elevation',
            zaxis=dict(range=[-1, 1])
        )
    )
    fig.show()

# Create widgets
size_w = widgets.IntSlider(min=20, max=200, step=10, value=100, description='Grid Size')
scale_w = widgets.FloatSlider(min=1.0, max=50.0, step=1.0, value=20.0, description='Noise Scale')
octaves_w = widgets.IntSlider(min=1, max=8, step=1, value=6, description='Octaves')
persistence_w = widgets.FloatSlider(min=0.1, max=1.0, step=0.05, value=0.5, description='Persistence')
lacunarity_w = widgets.FloatSlider(min=1.0, max=4.0, step=0.1, value=2.0, description='Lacunarity')
seed_w = widgets.IntSlider(min=0, max=100, step=1, value=42, description='Random Seed')
colorscale_w = widgets.Dropdown(
    options=['Viridis', 'Cividis', 'Earth', 'Terrain', 'Electric'],
    value='Terrain',
    description='Theme'
)

# Arrange layout
ui = widgets.VBox([
    widgets.HBox([size_w, scale_w]),
    widgets.HBox([octaves_w, persistence_w]),
    widgets.HBox([lacunarity_w, seed_w]),
    colorscale_w
])

out = widgets.interactive_output(update_plot, {
    'size': size_w, 'scale': scale_w, 'octaves': octaves_w, 
    'persistence': persistence_w, 'lacunarity': lacunarity_w, 
    'seed': seed_w, 'colorscale': colorscale_w
})

display(ui, out)
```
